# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df_insurance = pd.read_csv(url)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

- ganancias totales = ingresos mensuales * nº de meses contratados - gastos
- Calcular el min() de nº de meses contratados para hacer una comparación entre todas las filas de forma igualitaria.

```python
columnas_seleccionadas = ['monthly_premium_auto', 'months_since_policy_inception', 'sales_channel', 'total_claim_amount']
df_c2 = df_insurance[columnas_seleccionadas]
```

```python
df_c2.head()
```

```python
valor_minimo = df_c2['months_since_policy_inception'].min()
valor_minimo
```

```python
frecuencia_valores = df_c2['months_since_policy_inception'].value_counts()
frecuencia_valores
```

```python
df_c2.isnull().sum()
```

- Quiero obtener la duración del percentil 25 para calcular mi estrategia de marketing basada en los datos de contratación más recientes.

```python
percentil_25 = df_c2['months_since_policy_inception'].quantile(0.25)
percentil_25
```

- Filtro los clientes que contrataron hace 24 meses o antes.

```python
df_c2 = df_c2[df_c2['months_since_policy_inception'] <= percentil_25]
```

```python
df_c2.sample(10)
```

```python
df_c2.shape
```

- Tabla resumen con pivot.

- Ganancia para cada fila:

```python
df_c2['ganancia'] = df_c2['monthly_premium_auto'] * df_c2['months_since_policy_inception'] - df_c2['total_claim_amount']
```

- Tabla dinámica agrupada por 'sales_channel' y sumando la columna 'ganancia'

```python
df_resumen = pd.pivot_table(df_c2, values='ganancia', index='sales_channel', aggfunc='sum')
```

```python
df_resumen = df_resumen.round(2)
```

```python
df_resumen
```
